# File Transformation

This notebook has to be run first to transform Damirs ```DataTransformed.txt``` to ```reprocessed_data.pkl```. The ```preprocessed_data.pkl``` is the input-file for most other (following) scripts and notebooks!

````python
# TODO: add switzerland children and store in separat file
````

````python
import pandas as pd
import os
import pickle
from pandas.api.types import CategoricalDtype
````

````python
input_file = 'DataTransformed.csv'
input_folder = '../data/input/'
````

````python
data = pd.read_csv(os.path.join(input_folder, input_file))
````

The columntype for the 2 target columns will be changed to 'category' type:

````python
for col in ['diagnosis', 'location']:
    data[col] = data[col].astype('category')
````

````python
status_type = CategoricalDtype(categories=['AD_Urban', 'AD_Rural', 'HC_Urban', 'HC_Rural'], ordered=True)
data['diagnosis_location'] = data['diagnosis_location'].astype(status_type)
````

````python
# store data as pickle to output-folder
with open('../data/output/preprocessed_data.pkl','wb') as outfile:
    pickle.dump(data,outfile)
````
