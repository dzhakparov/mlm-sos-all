# Statistical Tests

````python
import pandas as pd
from pandas.api.types import CategoricalDtype
from statTests.statSuite import StatSuite
import warnings
import os, shutil

pd.set_option('display.max_rows', 500)
````

````python
input_file = 'DataTransformed.csv'
input_folder = '/home/schmidmarco/Documents/CODE/PROJECTS/sosall/data/input/'
output_folder = '/home/schmidmarco/Documents/CODE/PROJECTS/sosall/data/output/statisticalTests_2022-08-02'
````

This notebook calculates statistical tests (written in a own package called **'statSuite'**) of the **south african children** for the **questionary** in the SOSALL-Project.

Data have been preprocessed by Damir and have been taken without any changements in this notebook.

````python
data = pd.read_csv(os.path.join(input_folder, input_file))
````

The columntype for the 2 target columns will be changed to 'category' type:

````python
for col in ['diagnosis', 'location']:
    data[col] = data[col].astype('category')
````

````python
status_type = CategoricalDtype(categories=['AD_Urban', 'AD_Rural', 'HC_Urban', 'HC_Rural'], ordered=True)
data['diagnosis_location'] = data['diagnosis_location'].astype(status_type)
````

Running statistical analysis for ```diagnosis```, ```location``` and its combination ```diagnosis_location```. The warnings will be disabled because there are several warnings for shapiro-test.

> UserWarning: Input data for shapiro has range zero. The results may not be accurate.
warnings.warn("Input data for shapiro has range zero."

````python
try:
    os.mkdir(output_folder)
    print("Directory " , output_folder ,  " Created ")
except FileExistsError:
     for filename in os.listdir(output_folder):
        file_path = os.path.join(output_folder, filename)
        try:
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)
        except Exception as e:
            print('Failed to delete %s. Reason: %s' % (file_path, e))
````

````python
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for target in ['diagnosis', 'location', 'diagnosis_location']:
        obj = StatSuite(data=data, target=target, path=output_folder)
        obj.run(store_csv=True)
        # obj.plot(plot_choice=['histogram'])
        os.rename(os.path.join(output_folder, 'statisticalTests', 'output.csv'),
                  os.path.join(output_folder, 'statisticalTests', f'statistics_{target}.csv'))
        # os.rename(os.path.join(output_folder, 'statisticalTests','plot'),
        #          os.path.join(output_folder, 'statisticalTests', f'plot_{target}'))
````
