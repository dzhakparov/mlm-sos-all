Running statistical analysis for ```diagnosis```, ```location``` and its combination ```diagnosis_location```. The warnings will be disabled because there are several warnings for shapiro-test.

> UserWarning: Input data for shapiro has range zero. The results may not be accurate.
warnings.warn("Input data for shapiro has range zero."

````python
try:
    os.mkdir(config['output_folder'])
    print("Directory " , config['output_folder'] ,  " Created ")
except FileExistsError:
     for filename in os.listdir(config['output_folder']):
        file_path = os.path.join(config['output_folder'], filename)
        try:
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)
        except Exception as e:
            print('Failed to delete %s. Reason: %s' % (file_path, e))
````

````python
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for target in ['diagnosis', 'location', 'diagnosis_location']:
        obj = StatSuite(data=data, target=target, path=config['output_folder'])
        obj.run(store_csv=True)
        # obj.plot(plot_choice=['histogram'])
        os.rename(os.path.join(config['output_folder'], 'statisticalTests', 'output.csv'),
                  os.path.join(config['output_folder'], 'statisticalTests', f'statistics_{target}.csv'))
        # os.rename(os.path.join(output_folder, 'statisticalTests','plot'),
        #          os.path.join(output_folder, 'statisticalTests', f'plot_{target}'))
````
